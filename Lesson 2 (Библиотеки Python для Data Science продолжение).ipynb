### 1. В чём различие между зависимыми и независимыми выборками?

### 
При зависимых выборках применяется ранговый Т-критерий Уилкоксона. так же для зависимых выборок - тест Мак-Немара или Q-критерий Кохрена для сравнения трех и более групп. 
При независимых выборках группы должны быть независимыми, то есть критерий хи-квадрат не должен применяться при сравнении наблюдений "до-после" или связанных пар.

### Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

### 
Параметрический критерий – это метод статистического вывода, который применяется в отношении параметров генеральной совокупности на основе расчета их оценок по выборочным данным. При невозможности использования критерия Стьюдента для проверки гипотезы о равности мат, мы можем мы сможем воспользоваться его непараметрическим аналогом - критерием Манна-Уитни, который не требователен к закону распределения.
